```python
import csv
import ast
values = []

with open('sensor9y.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        # print(f'\t{row["sensor"]} works in the {row["9y"]}.')
        values.append(ast.literal_eval(row['9y']))
        line_count += 1
    print(f'Processed {line_count} lines.')
print(values)
```

```python

for i in range(1, len(values)-1):
    if abs(values[i]-values[i-1]) > 25 and abs(values[i]-values[i+1]) > 25:
        values[i] = round(((values[i-1]+values[i+1])/2), 3)
print(values)
```

```python
file1 = open("smooth9y2.csv", "a")  # append mode 
for v in values:
    file1.write(str(v)+",\n") 
file1.close()
```
