```python
import csv
import ast
allvalues = [[] for q in range(0, 50)]
c_names = ['0x', '0y', '1x', '1y', '2x', '2y', '3x', '3y', '4x', '4y', '5x', '5y', '6x', '6y', '7x', '7y', '8x', '8y', '9x', '9y', '10x', '10y', '11x', '11y', '12x', '12y', '13x', '13y', '14x', '14y', '15x', '15y', '16x', '16y', '17x', '17y', '18x', '18y', '19x', '19y', '20x', '20y', '21x', '21y', '22x', '22y', '23x', '23y', '24x', '24y']

with open('action1.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print([i for i in row])
            line_count += 1
        for j in range(0, 50):
            allvalues[j].append(ast.literal_eval(row[c_names[j]]))
        line_count += 1
    print(f'Processed {line_count} lines.')
print(allvalues[0])
```

```python
for values in allvalues:
    for i in range(1, len(values)-1):
        if abs(values[i]-values[i-1]) > 99 and abs(values[i]-values[i+1]) > 99:
            values[i] = round(((values[i-1]+values[i+1])/2), 3)
    print(values)
```

```python
file1 = open("smoothaction1.csv", "a")  # append mode 
for r in range(0,len(allvalues[0])):
    tow = ''
    for m in range(0, 50):
        tow += (str(allvalues[m][r])+",")
    tow = tow[:-1]
    tow += '\n'
    print(tow)
    file1.write(tow) 
file1.close()
```
